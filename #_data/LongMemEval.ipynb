```python
# Step 2: Prepare a subset of the dataset with 10 entries
import json
import os

# Load the dataset
dataset_path = 'LongMemEval/data/longmemeval_s_cleaned.json'
subset_path = 'LongMemEval/data/longmemeval_subset.json'

if not os.path.exists(dataset_path):
    raise FileNotFoundError(f"Dataset not found at {dataset_path}. Please ensure the dataset is downloaded.")

# Extract 10 entries
with open(dataset_path, 'r', encoding='utf-8') as f:  # Specify UTF-8 encoding
    data = json.load(f)
subset = data[:10]  # Take the first 10 entries

# Save the subset
with open(subset_path, 'w', encoding='utf-8') as f:  # Specify UTF-8 encoding
    json.dump(subset, f, indent=4)
print(f"Subset saved to {subset_path}")
```

```python
import os 
import json 

# result_data_path = "./../my_hypothesis.jsonl.eval-results-gpt40"
result_data_path = "./../my_hypothesis.jsonl.eval-results-gpt-4o"

if not os.path.exists(result_data_path):
    raise FileNotFoundError(f"Result data not found at {result_data_path}. Please ensure the result data is available.")

with open(result_data_path, 'r', encoding='utf-8') as f:  # Specify UTF-8 encoding
    result_data = [json.loads(line) for line in f]  # Parse each line as JSON
    # result_data
len(result_data)
```

```python
# Extract and print all question_ids
for entry in result_data:
    print(entry.get('question_id'), ": ", entry.get('autoeval_label').get('label'))
```

```python
count = 0
for entry in result_data:
    # print(type(entry.get('autoeval_label').get('label')))
    if entry.get('autoeval_label').get('label') == True:
        count += 1

print(f"Number of entries with autoeval_label 'True': {count}")
```

```python
percentage = (count / len(result_data)) * 100
print(f"Percentage of entries with autoeval_label 'True': {percentage:.2f}%")
```
