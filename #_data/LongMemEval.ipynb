```python
# Step 2: Prepare a subset of the dataset with 10 entries
import json
import os

# Load the dataset
dataset_path = 'LongMemEval/data/longmemeval_s_cleaned.json'
subset_path = 'LongMemEval/data/longmemeval_subset.json'

if not os.path.exists(dataset_path):
    raise FileNotFoundError(f"Dataset not found at {dataset_path}. Please ensure the dataset is downloaded.")

# Extract 10 entries
with open(dataset_path, 'r', encoding='utf-8') as f:  # Specify UTF-8 encoding
    data = json.load(f)
subset = data[:10]  # Take the first 10 entries

# Save the subset
with open(subset_path, 'w', encoding='utf-8') as f:  # Specify UTF-8 encoding
    json.dump(subset, f, indent=4)
print(f"Subset saved to {subset_path}")
```
